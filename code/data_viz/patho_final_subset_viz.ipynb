# Hotspot Viz

```python
hotspot_df = df[df["Hotspot"] == 1]
hotspot_df = hotspot_df.drop(columns=['Chr_Pos', 'Chr_SRA'])
print(hotspot_df.shape)
```

```python
hotspot_df.head()
```

```python
# --- Bar Plot 1: Top 20 Genes with Hotspots (Color Coded by OncoG/TSG) ---

# Calculate unique sample counts for each gene
unique_gene_counts = hotspot_df.groupby('Ref.Gene')['Sample_Name'].nunique()

# Calculate the total number of unique samples
total_unique_samples = hotspot_df['Sample_Name'].nunique()

# Corrected percentage representation based on total unique samples
corrected_unique_gene_percentage = (unique_gene_counts / total_unique_samples) * 100

# Select the top 20 genes based on unique sample counts
corrected_top_20_genes_unique = corrected_unique_gene_percentage.nlargest(20)
corrected_top_20_counts_unique = unique_gene_counts.loc[corrected_top_20_genes_unique.index]  # Actual unique sample counts

# Extract classification (Oncogene or TSG) for coloring
gene_classification = hotspot_df[['Ref.Gene', 'OncoG_TSG']].drop_duplicates().set_index('Ref.Gene')
gene_classification = gene_classification.loc[corrected_top_20_genes_unique.index]

# Define colors
colors = {'Oncogene': 'red', 'TSG': 'blue'}
bar_colors = [colors.get(gene_classification.loc[gene, 'OncoG_TSG'], 'gray') for gene in corrected_top_20_genes_unique.index]

# Plot
plt.figure(figsize=(12, 6))
bars = plt.bar(corrected_top_20_genes_unique.index, corrected_top_20_genes_unique.values, color=bar_colors)

# Annotate each bar with percentage
for bar, gene in zip(bars, corrected_top_20_genes_unique.index):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(),
             f'{bar.get_height():.1f}%', ha='center', va='bottom', fontsize=10)

# Modify x-tick labels: Gene name at 90 degrees
plt.xticks(range(len(corrected_top_20_genes_unique.index)), corrected_top_20_genes_unique.index, rotation=90, ha='center')

# Add sample counts separately below each bar
for bar, gene in zip(bars, corrected_top_20_genes_unique.index):
    plt.text(bar.get_x() + bar.get_width() / 2, -3, 
             f'({corrected_top_20_counts_unique[gene]})', ha='center', fontsize=10)

plt.ylabel('Unique Sample Count (%)')
plt.title('Top 20 Hotspot Pathogenic Genes by Unique Sample Frequency')

# Add grid in the background
plt.grid(axis='y', linestyle='--', linewidth=0.7, alpha=0.7)

# Legend for classification
plt.legend(handles=[plt.Line2D([0], [0], color='red', lw=4, label='Oncogene'),
                    plt.Line2D([0], [0], color='blue', lw=4, label='TSG')])

plt.ylim(-5, max(corrected_top_20_genes_unique.values) + 5)  # Adjust y-limits to accommodate labels

# Show plot
plt.show()

# Create a corrected table
table_hotspot_df = pd.DataFrame({
    'Gene': corrected_top_20_genes_unique.index,
    'Unique Sample Count': corrected_top_20_counts_unique.values,
    'Percentage (%)': corrected_top_20_genes_unique.values
})

print(table_hotspot_df)
```

```python
# --- Bar Plot 2: Hotspots by Chromosome with Stacked OncoG/TSG Bars ---

# Analyze frequency of hotspots by chromosome for OncoG/TSG
chromosome_hotspot_counts = hotspot_df[hotspot_df['Hotspot'] == 1].groupby(['CHROM_x', 'OncoG_TSG']).size().unstack(fill_value=0)

# Sort the index (chromosomes) naturally if possible
chromosome_hotspot_counts = chromosome_hotspot_counts.sort_index(key=lambda x: x.str.extract(r'(\d+|X|Y)')[0].map(lambda y: float(y) if y.isdigit() else y))

# Plot stacked bar chart
plt.figure(figsize=(10, 6))
chromosome_hotspot_counts.plot(kind='bar', stacked=True, title='Hotspots by Chromosome (OncoG vs TSG)', ylabel='Sample Count', color=['orange', 'green'])

# Customize plot
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# General Landscape

```python
landscape_df = df.drop(columns=['Chr_Pos', 'Chr_SRA'])
print(landscape_df.shape)
```

```python
# Count the number of samples per cancer type
sample_count_per_cancer = landscape_df["Cancer"].value_counts().reset_index()
sample_count_per_cancer.columns = ["Cancer", "Sample Count"]
print(sample_count_per_cancer)
```

```python
# --- Top Gene in each cancer ---#
# Drop rows with missing Cancer type
landscape_df_cancer = landscape_df.dropna(subset=["Cancer"])

# Count occurrences of each gene per cancer type
top_mutated_genes = landscape_df_cancer.groupby("Cancer")["Ref.Gene"].agg(lambda x: x.value_counts().idxmax())

# Convert to DataFrame for display
top_mutated_genes_df = top_mutated_genes.reset_index()
top_mutated_genes_df.columns = ["Cancer", "Top Mutated Gene"]
print(top_mutated_genes_df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Define the consistent order for the X-axis
cancer_order = ['AMPCa', 'HNSC', 'ESCA', 'LUAD', 'LUSC', 'NSCLC']

# Sample datasets for demonstration (replace with your data)
# Data for the bar plot
data_a = {
    'Cancer Type': ['AMPCa', 'HNSC', 'ESCA', 'LUAD', 'LUSC', 'NSCLC'],
    'Samples': [591, 328, 110, 1283, 802, 670]
}
df_a = pd.DataFrame(data_a)

# Data for the violin plot
data_b = {
    'Cancer Type': np.random.choice(['AMPCa', 'HNSC', 'ESCA', 'LUAD', 'LUSC', 'NSCLC'], 1000),
    'Mutation Percentage': np.random.rand(1000) * 100
}
df_b = pd.DataFrame(data_b)
df_b['Cancer Type'] = pd.Categorical(df_b['Cancer Type'], categories=cancer_order, ordered=True)

# Data for the heatmap
data_c = {
    'Cancer Type': ['AMPCa', 'HNSC', 'ESCA', 'LUAD', 'LUSC', 'NSCLC'],
    'TP53': [50, 80, 90, 60, 75, 55],
    'KRAS': [20, 10, 40, 70, 45, 35],
    'PIK3CA': [10, 15, 25, 30, 35, 20]
}
df_c = pd.DataFrame(data_c).set_index('Cancer Type')
df_c = df_c.reindex(cancer_order).transpose()  # Transpose the dataframe

# Plot 1: Bar Plot (Samples in Studies)
plt.figure(figsize=(12, 6))
sns.barplot(x='Cancer Type', y='Samples', data=df_a, palette='viridis', order=cancer_order)
plt.title('Samples in Studies', fontsize=16)
plt.xlabel('Cancer Type', fontsize=14)
plt.ylabel('# of Samples', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 2: Enhanced Violin Plot with Overlayed Points
plt.figure(figsize=(16, 8))

# Create a violin plot for distributions
sns.violinplot(
    x='Cancer Type',
    y='Mutation Percentage',
    data=df_b,
    order=cancer_order,
    palette='muted',
    inner=None,  # Removes inner elements for clarity
)

# Overlay individual points (scatter plot)
sns.stripplot(
    x='Cancer Type',
    y='Mutation Percentage',
    data=df_b,
    order=cancer_order,
    color='black',  # Scatter points as black dots
    size=3,  # Control dot size
    alpha=0.6,  # Add some transparency
    jitter=True  # Adds horizontal jitter to avoid overlap
)

# Add titles and labels
plt.title('Mutations in Cancer Type', fontsize=16)
plt.xlabel('Cancer Type', fontsize=14)
plt.ylabel('% of Samples with a Mutant Gene', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Plot 3: Heatmap (Mutation Frequency with Cancer on X-axis)
plt.figure(figsize=(12, 6))
sns.heatmap(df_c, annot=True, fmt='d', cmap='Blues', cbar_kws={'label': 'Frequency'})
plt.title('Mutation Frequency by Gene', fontsize=16)
plt.xlabel('Cancer Type', fontsize=14)
plt.ylabel('Gene', fontsize=14)
plt.tight_layout()
plt.show()
```
